```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

class DhakaTrafficDatasetGenerator:
    """Generate synthetic traffic dataset for Dhaka city"""

    def __init__(self):
        # Major intersections in Dhaka
        self.dhaka_intersections = [
            {"name": "Shahbag", "lat": 23.7389, "lon": 90.3944, "type": "edge"},
            {"name": "Farmgate", "lat": 23.7574, "lon": 90.3890, "type": "edge"},
            {"name": "Karwan Bazar", "lat": 23.7508, "lon": 90.3915, "type": "non_edge"},
            {"name": "Mohakhali", "lat": 23.7808, "lon": 90.4067, "type": "edge"},
            {"name": "Banani", "lat": 23.7937, "lon": 90.4066, "type": "non_edge"},
            {"name": "Gulshan 1", "lat": 23.7805, "lon": 90.4161, "type": "non_edge"},
            {"name": "Gulshan 2", "lat": 23.7925, "lon": 90.4078, "type": "non_edge"},
            {"name": "Badda", "lat": 23.7808, "lon": 90.4264, "type": "edge"},
            {"name": "Rampura", "lat": 23.7636, "lon": 90.4253, "type": "edge"},
            {"name": "Malibagh", "lat": 23.7461, "lon": 90.4065, "type": "non_edge"},
            {"name": "Mouchak", "lat": 23.7474, "lon": 90.4110, "type": "non_edge"},
            {"name": "Moghbazar", "lat": 23.7518, "lon": 90.4047, "type": "non_edge"},
            {"name": "Shantinagar", "lat": 23.7379, "lon": 90.4073, "type": "non_edge"},
            {"name": "Kakrail", "lat": 23.7344, "lon": 90.4023, "type": "non_edge"},
            {"name": "Paltan", "lat": 23.7341, "lon": 90.4112, "type": "edge"},
            {"name": "Motijheel", "lat": 23.7308, "lon": 90.4168, "type": "edge"},
            {"name": "Dhanmondi 27", "lat": 23.7461, "lon": 90.3742, "type": "non_edge"},
            {"name": "Dhanmondi 32", "lat": 23.7544, "lon": 90.3754, "type": "non_edge"},
            {"name": "Kalabagan", "lat": 23.7408, "lon": 90.3814, "type": "non_edge"},
            {"name": "Science Lab", "lat": 23.7429, "lon": 90.3876, "type": "non_edge"},
            {"name": "Nilkhet", "lat": 23.7354, "lon": 90.3840, "type": "non_edge"},
            {"name": "New Market", "lat": 23.7341, "lon": 90.3888, "type": "non_edge"},
            {"name": "Azimpur", "lat": 23.7284, "lon": 90.3844, "type": "edge"},
            {"name": "Jatrabari", "lat": 23.7099, "lon": 90.4326, "type": "edge"},
            {"name": "Sayedabad", "lat": 23.7156, "lon": 90.4279, "type": "edge"},
            {"name": "Malibagh Railgate", "lat": 23.7501, "lon": 90.4153, "type": "non_edge"},
            {"name": "Tejgaon", "lat": 23.7644, "lon": 90.3967, "type": "edge"},
            {"name": "Mohammadpur", "lat": 23.7654, "lon": 90.3564, "type": "edge"},
            {"name": "Asad Gate", "lat": 23.7609, "lon": 90.3682, "type": "non_edge"},
            {"name": "Russell Square", "lat": 23.7522, "lon": 90.3783, "type": "non_edge"},
            {"name": "Panthpath", "lat": 23.7532, "lon": 90.3851, "type": "non_edge"},
            {"name": "Sonargaon", "lat": 23.7543, "lon": 90.3934, "type": "non_edge"},
            {"name": "Bijoy Sarani", "lat": 23.7673, "lon": 90.3905, "type": "non_edge"},
            {"name": "Agargaon", "lat": 23.7757, "lon": 90.3798, "type": "edge"},
            {"name": "Mirpur 1", "lat": 23.7956, "lon": 90.3537, "type": "edge"},
            {"name": "Mirpur 10", "lat": 23.8069, "lon": 90.3687, "type": "edge"},
            {"name": "Mirpur 11", "lat": 23.8163, "lon": 90.3779, "type": "edge"},
            {"name": "Kazipara", "lat": 23.7934, "lon": 90.3714, "type": "non_edge"}
        ]

        # Road connections (from_id, to_id, road_name, distance_km, lanes)
        self.road_connections = [
            (0, 19, "Elephant Road", 1.2, 4), (19, 2, "Karwan Bazar Road", 0.8, 4),
            (2, 1, "Tejgaon Link Road", 1.0, 4), (1, 26, "Farmgate-Tejgaon Road", 0.7, 4),
            (26, 3, "Mohakhali Link", 1.5, 4), (3, 4, "Banani Road", 1.8, 3),
            (4, 5, "Gulshan Avenue", 1.0, 3), (5, 6, "Gulshan Link Road", 1.2, 3),
            (6, 7, "Badda Link", 2.0, 3), (7, 8, "Rampura Road", 1.5, 3),
            (8, 9, "Malibagh Road", 1.8, 4), (9, 10, "Mouchak-Malibagh", 0.5, 4),
            (10, 11, "Moghbazar Road", 0.6, 4), (11, 0, "Shahbag Link", 1.0, 4),
            (0, 12, "Bangla Motor Road", 0.8, 3), (12, 13, "Shantinagar-Kakrail", 0.6, 3),
            (13, 14, "Paltan Road", 0.7, 4), (14, 15, "Motijheel Link", 0.8, 4),
            (15, 23, "Jatrabari Road", 3.5, 4), (23, 24, "Sayedabad Link", 1.0, 3),
            (0, 16, "Dhanmondi 27 Road", 1.5, 3), (16, 17, "Dhanmondi Internal", 0.8, 2),
            (17, 18, "Kalabagan Road", 0.7, 3), (18, 19, "Science Lab Link", 0.5, 3),
            (19, 20, "Nilkhet Road", 0.6, 3), (20, 21, "New Market Road", 0.5, 3),
            (21, 22, "Azimpur Road", 1.0, 3), (1, 29, "Russell Square Road", 0.6, 4),
            (29, 30, "Panthpath Road", 0.8, 4), (30, 31, "Sonargaon Road", 0.5, 4),
            (31, 2, "Karwan Bazar Link", 0.4, 4), (1, 32, "Bijoy Sarani Road", 1.2, 4),
            (32, 33, "Agargaon Link", 1.5, 4), (33, 34, "Mirpur Road Section 1", 2.5, 4),
            (34, 35, "Mirpur Road Section 2", 1.8, 4), (35, 36, "Mirpur Road Section 3", 1.5, 3),
            (36, 37, "Kazipara Link", 0.8, 3), (9, 25, "Railgate Road", 0.6, 3),
            (25, 10, "Mouchak Link", 0.5, 3), (27, 28, "Asad Gate Road", 1.2, 3),
            (28, 29, "Russell Square Link", 0.9, 3), (27, 17, "Mohammadpur-Dhanmondi", 1.5, 3),
            # Add reverse directions for bidirectional roads
            (19, 0, "Elephant Road", 1.2, 4), (2, 19, "Karwan Bazar Road", 0.8, 4),
            (1, 2, "Tejgaon Link Road", 1.0, 4), (26, 1, "Farmgate-Tejgaon Road", 0.7, 4),
            (3, 26, "Mohakhali Link", 1.5, 4), (4, 3, "Banani Road", 1.8, 3),
            (5, 4, "Gulshan Avenue", 1.0, 3), (6, 5, "Gulshan Link Road", 1.2, 3),
            (7, 6, "Badda Link", 2.0, 3), (8, 7, "Rampura Road", 1.5, 3),
            (9, 8, "Malibagh Road", 1.8, 4), (10, 9, "Mouchak-Malibagh", 0.5, 4),
            (11, 10, "Moghbazar Road", 0.6, 4), (0, 11, "Shahbag Link", 1.0, 4),
            (12, 0, "Bangla Motor Road", 0.8, 3), (13, 12, "Shantinagar-Kakrail", 0.6, 3),
        ]

        self.vehicle_length = 5.0  # Average vehicle length in meters

    def generate_intersections(self):
        """Generate intersections dataset"""
        data = []
        for i, intersection in enumerate(self.dhaka_intersections):
            data.append({
                'intersection_id': i,
                'name': intersection['name'],
                'latitude': intersection['lat'],
                'longitude': intersection['lon'],
                'type': intersection['type'],
                'is_traffic_light': random.choice([True, False]),
                'average_wait_time': np.random.uniform(20, 90)
            })
        return pd.DataFrame(data)

    def generate_roads(self):
        """Generate roads dataset"""
        data = []
        for i, (from_id, to_id, road_name, distance_km, lanes) in enumerate(self.road_connections):
            data.append({
                'road_id': i,
                'road_name': road_name,
                'from_intersection': from_id,
                'to_intersection': to_id,
                'length_meters': distance_km * 1000,
                'num_lanes': lanes,
                'is_bidirectional': True,
                'road_type': 'main' if lanes >= 4 else 'secondary',
                'speed_limit': 40.0 if lanes >= 4 else 30.0,
                'has_bus_lane': lanes >= 4 and random.random() > 0.7
            })
        return pd.DataFrame(data)

    def generate_traffic_flow(self, roads_df, num_days=7):
        """Generate traffic flow dataset for specified number of days"""
        data = []
        start_date = datetime(2024, 1, 15, 0, 0, 0)

        for day in range(num_days):
            current_date = start_date + timedelta(days=day)
            day_name = current_date.strftime('%A')
            is_weekend = day_name in ['Friday', 'Saturday']

            for hour in range(24):
                timestamp = current_date + timedelta(hours=hour)

                # Determine if peak hour
                is_peak = (7 <= hour <= 10) or (17 <= hour <= 20)

                # Base traffic multiplier
                if is_weekend:
                    base_multiplier = 0.6
                else:
                    base_multiplier = 1.0

                # Peak hour multiplier
                if is_peak:
                    traffic_multiplier = base_multiplier * np.random.uniform(2.0, 3.5)
                elif 11 <= hour <= 16:
                    traffic_multiplier = base_multiplier * np.random.uniform(1.2, 1.8)
                elif 22 <= hour or hour <= 5:
                    traffic_multiplier = base_multiplier * np.random.uniform(0.2, 0.5)
                else:
                    traffic_multiplier = base_multiplier * np.random.uniform(0.8, 1.5)

                for _, road in roads_df.iterrows():
                    # Base traffic volume depends on number of lanes
                    base_traffic = road['num_lanes'] * 200

                    # Calculate inbound and outbound traffic
                    inbound = int(base_traffic * traffic_multiplier * np.random.uniform(0.8, 1.2))
                    outbound = int(inbound * np.random.uniform(0.7, 1.1))

                    # Calculate average speed (inversely related to congestion)
                    if traffic_multiplier > 2.5:
                        avg_speed = np.random.uniform(8, 15)
                    elif traffic_multiplier > 1.5:
                        avg_speed = np.random.uniform(15, 25)
                    else:
                        avg_speed = np.random.uniform(25, road['speed_limit'])

                    vehicle_density = inbound / (road['length_meters'] / 1000)

                    data.append({
                        'timestamp': timestamp,
                        'road_id': road['road_id'],
                        'inbound_traffic': inbound,
                        'outbound_traffic': outbound,
                        'average_speed': avg_speed,
                        'vehicle_density': vehicle_density,
                        'hour_of_day': hour,
                        'day_of_week': day_name,
                        'is_peak_hour': is_peak
                    })

        return pd.DataFrame(data)

    def calculate_congestion(self, traffic_df, roads_df):
        """Calculate congestion data based on paper formulas"""
        data = []

        # Group by road_id and sort by timestamp
        for road_id in traffic_df['road_id'].unique():
            road_data = traffic_df[traffic_df['road_id'] == road_id].sort_values('timestamp')
            road_info = roads_df[roads_df['road_id'] == road_id].iloc[0]

            li = road_info['num_lanes']
            di = road_info['length_meters']
            L = self.vehicle_length

            prev_vehicles = 0

            for idx, row in road_data.iterrows():
                fini = row['inbound_traffic']
                fouti = row['outbound_traffic']

                # Calculate fi(t-1) + fini(t) - fouti(t)
                net_vehicles = prev_vehicles + fini - fouti

                # Calculate congestion coefficient Ci(t) - Formula (8)
                if net_vehicles > 0:
                    Ci = (net_vehicles * L) / (li * di)
                else:
                    Ci = 0

                # Clip congestion coefficient to reasonable range
                Ci = min(Ci, 2.0)

                # Calculate road condition factor Ri(t) - Formula (9)
                Ri = di * (1 + Ci)

                # Determine congestion level (0-4)
                if Ci < 0.2:
                    congestion_level = 0  # Unobstructed
                elif Ci < 0.5:
                    congestion_level = 1  # Basically unblocked
                elif Ci < 0.8:
                    congestion_level = 2  # Mild congestion
                elif Ci < 1.2:
                    congestion_level = 3  # Moderate congestion
                else:
                    congestion_level = 4  # Severe congestion

                # Calculate travel times
                expected_time = (di / 1000) / (road_info['speed_limit']) * 60  # minutes
                actual_time = (di / 1000) / max(row['average_speed'], 1) * 60  # minutes

                data.append({
                    'timestamp': row['timestamp'],
                    'road_id': road_id,
                    'congestion_level': congestion_level,
                    'congestion_coefficient': round(Ci, 4),
                    'road_condition_factor': round(Ri, 2),
                    'travel_time_minutes': round(actual_time, 2),
                    'expected_travel_time': round(expected_time, 2)
                })

                # Update vehicles for next iteration
                prev_vehicles = max(0, net_vehicles)

        return pd.DataFrame(data)

    def generate_time_patterns(self, congestion_df, roads_df):
        """Generate time-based pattern statistics"""
        data = []

        for road_id in roads_df['road_id'].unique():
            road_congestion = congestion_df[congestion_df['road_id'] == road_id].copy()
            road_congestion['hour'] = pd.to_datetime(road_congestion['timestamp']).dt.hour
            road_congestion['day_type'] = pd.to_datetime(road_congestion['timestamp']).dt.day_name().apply(
                lambda x: 'weekend' if x in ['Friday', 'Saturday'] else 'weekday'
            )

            for hour in range(24):
                for day_type in ['weekday', 'weekend']:
                    subset = road_congestion[
                        (road_congestion['hour'] == hour) &
                        (road_congestion['day_type'] == day_type)
                    ]

                    if len(subset) > 0:
                        data.append({
                            'road_id': road_id,
                            'hour': hour,
                            'day_type': day_type,
                            'avg_congestion': round(subset['congestion_coefficient'].mean(), 4),
                            'std_congestion': round(subset['congestion_coefficient'].std(), 4),
                            'traffic_multiplier': round(subset['congestion_coefficient'].mean() + 1, 2)
                        })

        return pd.DataFrame(data)

    def generate_all_datasets(self, num_days=7, output_dir='./'):
        """Generate all datasets and save to CSV"""
        print("Generating Dhaka Traffic Dataset...")

        # Generate datasets
        print("1. Generating intersections...")
        intersections_df = self.generate_intersections()

        print("2. Generating roads...")
        roads_df = self.generate_roads()

        print(f"3. Generating traffic flow data for {num_days} days...")
        traffic_df = self.generate_traffic_flow(roads_df, num_days)

        print("4. Calculating congestion metrics...")
        congestion_df = self.calculate_congestion(traffic_df, roads_df)

        print("5. Generating time patterns...")
        patterns_df = self.generate_time_patterns(congestion_df, roads_df)

        # Save to CSV
        print(f"\nSaving datasets to {output_dir}...")
        intersections_df.to_csv(f'{output_dir}intersections.csv', index=False)
        roads_df.to_csv(f'{output_dir}roads.csv', index=False)
        traffic_df.to_csv(f'{output_dir}traffic_flow.csv', index=False)
        congestion_df.to_csv(f'{output_dir}congestion_data.csv', index=False)
        patterns_df.to_csv(f'{output_dir}time_patterns.csv', index=False)

        # Print summary statistics
        print("\n" + "="*60)
        print("DATASET GENERATION COMPLETE!")
        print("="*60)
        print(f"Intersections: {len(intersections_df)} records")
        print(f"Roads: {len(roads_df)} records")
        print(f"Traffic Flow: {len(traffic_df)} records")
        print(f"Congestion Data: {len(congestion_df)} records")
        print(f"Time Patterns: {len(patterns_df)} records")
        print("\nCongestion Level Distribution:")
        print(congestion_df['congestion_level'].value_counts().sort_index())
        print("\nAverage Congestion by Hour (Weekdays):")
        weekday_patterns = patterns_df[patterns_df['day_type'] == 'weekday']
        hourly_avg = weekday_patterns.groupby('hour')['avg_congestion'].mean()
        print(hourly_avg.round(3))

        return {
            'intersections': intersections_df,
            'roads': roads_df,
            'traffic_flow': traffic_df,
            'congestion_data': congestion_df,
            'time_patterns': patterns_df
        }


# Usage Example
if __name__ == "__main__":
    generator = DhakaTrafficDatasetGenerator()
    datasets = generator.generate_all_datasets(num_days=7, output_dir='./')

    print("\n✅ All datasets generated successfully!")
    print("Files created:")
    print("  - intersections.csv")
    print("  - roads.csv")
    print("  - traffic_flow.csv")
    print("  - congestion_data.csv")
    print("  - time_patterns.csv")
```
